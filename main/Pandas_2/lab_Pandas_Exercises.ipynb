1. Import the necessary libraries

2. In this exercise, we will use the [drink dataset](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv).

3. Assign it to a variable called drinks.

```python
drinks = pd.read_csv('C:\\Users\\uchid\OneDrive\\Documents\\Data Science MSc\\Modules\\Sem 1\\Programming for Data Science\\DataScienceProjects22\\main\\Pandas_2\\data\\drinks.csv')
drinks
```

4. Which continent drinks more beer on average?

```python
drinks.groupby('continent').beer_servings.mean()
# The EU.
```

5. For each continent print the statistics for spirit consumption.

```python
drinks.groupby('continent').spirit_servings.describe()
```

6. Print the mean alcohol consumption per continent for every column

```python
print(drinks.groupby('continent').mean())
```

7. Find out the top 3 continents in terms of beer consumption.

```python
# sum then sort from highest to lowest
beer_consumption = drinks.groupby('continent').beer_servings.sum().sort_values(ascending=False)

# return top 3
beer_consumption[0:3]
```

8. Add a column to drink called `total_servings_of_beer_and_wine` calculate the total_servings_of_beer_and_wine (beer_servings + wine_servings) from corresponding servings. Please check the numpy array operations in the Numpy lecture.

```python
drinks['total_servings_of_beer_and_wine'] = drinks['beer_servings'] + drinks['wine_servings']
drinks
```

9. Create a bar chart with the 5 continents that have the most beer consumption. Please add suitable x label, y label, and title.

```python
import matplotlib.pyplot as plt

# clean missing values 
drinks.fillna(0)

# group by the continent
continents = drinks.groupby('continent').sum()

# create the plot
continents['beer_servings'].plot(kind='bar')

# Set the title and labels
plt.xlabel('continents')
plt.ylabel('beer_servings')
plt.title('5 continents with the most beer consumption')

# show the plot
plt.show()
```

10. Create a subplot (with 3 scatterplots) with the beer_servings, spirit_servings, wine_servings by 8 countries which have most  total_servings_of_beer_and_wine. Please check the parameters **subplots** and **layout** in [**pd.plot()**](https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.DataFrame.plot.html) to make a subplot.

```python
# sorts by total_servings_of_beer_and_wine and gets first 8 rows
drinks2 = drinks.sort_values(by= 'total_servings_of_beer_and_wine', ascending=False)[0:8]

# first subplot
drinks2.plot(x= 'country', y= 'beer_servings', kind= 'scatter', subplots=True, layout= (3, 1))

# second subplot
drinks2.plot(x= 'country', y= 'spirit_servings', kind= 'scatter', subplots=True, layout= (3, 1))

# third subplot
drinks2.plot(x= 'country', y= 'wine_servings', kind= 'scatter', subplots=True, layout= (3, 1))

plt.show()
```
