3. Write a Python program to creat a Dictionary and get its all keys and values.

```python
# please write your code below and press Shift+Enter to see the result:

# create dictionary 
d = {}

# init keys and values
keysInsert = ['one', 'two', 'three', 'four', 'five']

for i in range(0, len(keysInsert)):
    d[keysInsert[i]] = i + 1

print(d)

# get all keys and values
print(d.items())

# gets keys
print(d.keys()) 

# gets values 
print(d.values())
```
